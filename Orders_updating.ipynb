```python
import pandas as pd
import numpy as np
import time
import datetime
import os
```

```python
# my functions
def four_column_union(l):
    if pd.isna(l[0]) == False:
        return l[0]
    elif pd.isna(l[1]) == False:
        return l[1]
    elif pd.isna(l[2]) == False:
        return l[2]
    else:
        return l[3]
    
    
def three_column_union(l):
    if pd.isna(l[0]) == False:
        return l[0]
    elif pd.isna(l[1]) == False:
        return l[1]
    else:
        return l[2]
    
def VIN_delete_spaces(l):
    if pd.isna(l) or l.strip() != '':
        return l
    else:
        return np.nan

    
def ADR(l):
    if pd.isna(l):
        return l
    elif l.find('E5V') != -1:
        return 'E5V'
    elif l.find('E5G') != -1:
        return 'E5G'
    elif l.find('C5A') != -1:
        return 'C5A'
    else:
        return np.nan
    

def Mega_cab(l):
    if pd.isna(l):
        return l
    elif l.find('F15') != -1:
        return 'F15'
    else:
        return np.nan
    
    
def baumuster(l):
    if pd.isna(l[1]) == False:
        return l[1]
    elif pd.isna(l[0]):
        return l[1]
    else:
        return f'{l[0][3:10]}2'
    

def status(l):
    if pd.isna(l[0]) == False:
        return 'Sold'
    elif pd.isna(l[1]):
        return 'Order'
    else:
        return 'Stock'
    
    
def ID_check(l):
    if pd.isna(l):
        return l
    elif l.find('D0Y') != -1 and l.find('F9R') != -1 and l.find('U8D') != -1 and l.find('U8Z') != -1:
        return 'CiW'
    else:
        return np.nan
    
    
def SC(l):
    if pd.isna(l):
        return l
    elif l[0] == '0':
        return l[1]
    else:
        return l[:1]
    
    
def Dep1C(l):
    if pd.isna(l[0]):
        return l[5]
    elif (l[0]=='xxx' and pd.isna(l[1]) == False and 
          datetime.date(int(l[1][6:]),int(l[1][3:5]),int(l[1][:2]))<datetime.date(2021,4,1)):
        return 'TS'
    elif l[2] == 'Unimog' or l[2] == 'Zetros':
        return 'ST'
    elif l[0] == 'yyy':
        if pd.isna(l[3]) == False:
            if datetime.date(int(l[1][6:]),int(l[1][3:5]),int(l[1][:2]))<datetime.date(2018,10,1):
                return 'BB'
            elif pd.isna(l[4]):
                return l[5]
            else:
                return l[4]
        elif (l[7] == 'Stock 01.01.18' or l[7] == '2017' or 
              datetime.date(l[6].year,l[6].month,l[6].day)<datetime.date(2018,10,1)):
            return 'BB'
        elif pd.isna(l[4]):
            return l[5]
        else:
            return l[4]
    elif pd.isna(l[4]):
        return l[5]
    else:
        return l[4]
            

        
def cab_code(l):
    if pd.isna(l):
        return l
    else:
        for code in ['F1A', 'F1B', 'F1D', 'F1E', 'F1H', 'F1H', 'F1O', 'F1S', 'F1T', 'F1W', 'F1F', 'F1I', 'F1J', 'F15', 'F1C', 
                     'F1X', 'F1Z', 'F07', 'F05', 'F04', 'F0E', 'F1Q']:
            if l.find(code) != -1:
                return code
        return np.nan
    
    
def fin_subs(l):
    if pd.isna(l[0]):
        if pd.isna(l[1]):
            return np.nan
        elif str(l[1]) == '0' or str(l[1]) == '1000' or str(l[1]) == '3000' or str(l[1]) == '3600' or str(l[1]) == '4000':
            return 'Regular'
        elif str(l[1]) == '6000' or str(l[1]) == '7500':
            return 'SPR'
        else:
            return l[1]
    elif str(l[0]) == '0' or str(l[0]) == '1000' or str(l[0]) == '3000' or str(l[0]) == '3600' or str(l[0]) == '4000':
        return 'Regular'
    elif str(l[0]) == '6000' or str(l[0]) == '7500':
        return 'SPR'
    else:
        return l[0]
    
    
def tyres_r(l):
    if pd.isna(l):
        return l
    else:
        d = {
            'R1I':22.5,
            'R3Q':24,
            'R3P':24
        }
        for code in list(d.keys()):
            if l.find(code) != -1:
                return d[code]
        return np.nan
    
def turnover(l):
    if pd.isna(l[0]):
        return np.nan
    elif pd.isna(l[1]) == False:
        end = l[0].split('.')
        start = l[1].split('.')
        return str(datetime.date(int(end[2]), int(end[1]), int(end[0])) - datetime.date(int(start[2]), int(start[1]), 
                                                                                        int(start[0]))).split()[0]
    elif pd.isna(l[2]) == False:
        end = l[0].split('.')
        start = l[2].split('.')
        return str(datetime.date(int(end[2]), int(end[1]), int(end[0])) - datetime.date(int(start[2]), int(start[1]), 
                                                                                        int(start[0]))).split()[0]
    elif pd.isna(l[3]) == False:
        end = l[0].split('.')
        start = l[3].split('.')
        return str(datetime.date(int(end[2]), int(end[1]), int(end[0])) - datetime.date(int(start[2]), int(start[1]), 
                                                                                        int(start[0]))).split()[0]
    else:
        return np.nan
    
    
def segments(l):
    if l[0] == 'Unimog':
        return 'HDT (Others)'
    elif pd.isna(l[1]) == False:
        return l[1]
    elif pd.isna(l[2]) == False:
        return l[2]
    elif pd.isna(l[3]) == False:
        return l[3]
    elif pd.isna(l[4]) == False:
        return l[4]
    else:
        return l[5]
    
    
def segments2(l):
    if pd.isna(l[0]) == False and (l[0] == 'HDT/A 4x2 (Actros premium)' or l[0] == 'HDT/A 4x2 (Actros budget)'):
        return 'HDT/A 4x2 (Actros)'
    elif pd.isna(l[1]) == False:
        return l[1]
    elif pd.isna(l[2]) == False:
        return l[2]
    else:
        return l[0]
    
```

# 3. Создание списка заказов

```python
%%time

# Из сводной по файлу Проценко и стока на 2018 формируется список заказов

bad_packs = {
    '1962462006 - 009':[1952462006,1952462009],
    '18524300019 - 118':[1852430019,1852430118],
    '182430928-929':[1852430928, 1852430929]
}

res = []
num = '          '
end_num = '          '
for order in stock_18['Order']:
    res.append([int(order), str(int(order))])

for pack in pivot['Order ']:
    for order in pack:
        if order in bad_packs.keys():
            print(order)
            num, end_num = bad_packs[order]
            
        elif order.find('-') == -1:
            num = int(order)
            end_num = int(order)
        else:
            num = order[:order.find('-')].strip()
            num = int(str(end_num)[:-len(num)] + num)
            end_num = order[order.find('-') + 1:].strip()
            end_num = int(str(num)[:-len(end_num)] + end_num)
        if num > end_num:
            end_num += int('1' + '0' * len(order[order.find('-') + 1:].strip()))
        for i in range(num, end_num + 1):
            res.append([i, ', '.join(pack)])
            
pivot['Order '] = pivot['Order '].apply(lambda x: ', '.join(x))
out = pd.DataFrame(res, columns = ['Order', 'initial'])

out = out[~out['initial'].isin(list(pivot[pivot['Итог'] == 0]['Order '].values))]
```
